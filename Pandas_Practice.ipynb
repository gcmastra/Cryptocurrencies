```python
import pandas as pd
file_path = "Resources/iris.csv"
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# section 18.2.2 says unsupervised learning only uses numerical data so we can drop the class column
new_iris_df = iris_df.drop(["class"], axis=1)
new_iris_df.head()
```

```python
# save the new dataframe back out to a CSV to reuse later and in case we ever need to restore from baseline
output_file_path = "Resources/new_iris_data.csv"
new_iris_df.to_csv(output_file_path, index=False)
```

```python
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path)
df_shopping.head()
```

```python
df_shopping.dtypes
```

```python
# there appear to be two rows missing age and one row missing spending score, or possibly some rows missing all 4
df_shopping.describe()
```

```python
df_shopping = df_shopping.dropna()
```

```python
df_shopping.drop(columns=["CustomerID"], inplace=True)
df_shopping.describe()
# this is end of section 18.2.4
```

```python
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
    
df_shopping["Card Member"] = df_shopping["Card Member"].apply(change_string)
df_shopping.describe()
```
